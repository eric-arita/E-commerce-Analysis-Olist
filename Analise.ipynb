```python
import pandas as pd
import os
```

```python


base_path = "00. Data"

arquivos = [
    "olist_customers_dataset.csv",
    "olist_geolocation_dataset.csv",
    "olist_order_items_dataset.csv",
    "olist_order_payments_dataset.csv",
    "olist_order_reviews_dataset.csv",
    "olist_orders_dataset.csv",
    "olist_products_dataset.csv",
    "olist_sellers_dataset.csv",
    "product_category_name_translation.csv"
]

dfs = {}

for arquivo in arquivos:
    nome_df =  arquivo.replace(".csv", "")
    path = os.path.join(base_path, arquivo)
    dfs[nome_df] = pd.read_csv(path)

# df_customers = dfs["olist_customers_dataset"]
# df_geolocation =   dfs["olist_customers_dataset"]
```

```python
df_1 = pd.merge(dfs["olist_order_items_dataset"], dfs["olist_products_dataset"][['product_id','product_category_name']],on='product_id', how='left')
df_1
```

```python
df_1[df_1['order_item_id']>1]
```

```python
df_2 = df_1.loc[df_1.groupby(['order_id', 'product_id'])['order_item_id'].idxmax()]
df_2
```

```python
df_2['P*Q'] = df_2['order_item_id']  * df_2['price']
df_2.sort_values('P*Q', ascending=False).head(30)
```

```python
top5_category = list(df_2.groupby(['product_category_name'])['P*Q'].sum().reset_index().sort_values(by='P*Q', ascending= False).head(5)['product_category_name'])
top5_category
```

```python
df_3 = df_2[df_2['product_category_name'].isin(top5_category)].sort_values(by='P*Q',ascending=False).reset_index(drop=True)
df_3
```

```python
df_3['product_id'].value_counts().head(20)
```

```python
df_3 = df_3.groupby(['product_category_name','product_id','price']).sum().reset_index()
df_3
```
